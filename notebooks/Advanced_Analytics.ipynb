```python
import re

raw_dir = os.path.join(project_path, "data", "raw")

# only keep files matching 01_ through 10_ with no "(1)" duplicate suffix
keep_pattern = re.compile(r"^(0[1-9]|10)_.*\.csv$")

removed = []
kept = []

for fname in os.listdir(raw_dir):
    fpath = os.path.join(raw_dir, fname)
    if keep_pattern.match(fname):
        kept.append(fname)
    else:
        os.remove(fpath)
        removed.append(fname)

print(f"Kept {len(kept)} file(s):")
for f in sorted(kept):
    print(" -", f)

print(f"\nRemoved {len(removed)} file(s):")
for f in sorted(removed):
    print(" -", f)
```

```python
RAW_DIR = os.path.join("data", "raw")

FILES = {
    "fund_master":           "01_fund_master.csv",
    "nav_history":            "02_nav_history.csv",
    "aum_by_fund_house":       "03_aum_by_fund_house.csv",
    "monthly_sip_inflows":      "04_monthly_sip_inflows.csv",
    "category_inflows":          "05_category_inflows.csv",
    "industry_folio_count":       "06_industry_folio_count.csv",
    "scheme_performance":          "07_scheme_performance.csv",
    "investor_transactions":        "08_investor_transactions.csv",
    "portfolio_holdings":             "09_portfolio_holdings.csv",
    "benchmark_indices":                "10_benchmark_indices.csv",
}

dfs = {}
for name, filename in FILES.items():
    path = os.path.join(RAW_DIR, filename)
    dfs[name] = pd.read_csv(path)

print("Loaded", len(dfs), "datasets:", list(dfs.keys()))
```

```python
def profile(name, df):
    print("=" * 90)
    print(f"DATASET: {name}")
    print("=" * 90)
    print("Shape:", df.shape)
    print("\nDtypes:")
    print(df.dtypes)
    print("\nHead:")
    display(df.head())
    n_dupes = df.duplicated().sum()
    n_nulls = int(df.isna().sum().sum())
    print(f"Duplicate rows: {n_dupes} | Total null cells: {n_nulls}")
    if n_nulls:
        nc = df.isna().sum()
        print("Nulls by column (non-zero):")
        print(nc[nc > 0])
    print()

for name, df in dfs.items():
    profile(name, df)
```

```python
import time

SCHEMES = {
    "HDFC_Top_100_Direct": 125497,
    "SBI_Bluechip": 119551,
    "ICICI_Bluechip": 120503,
    "Nippon_Large_Cap": 118632,
    "Axis_Bluechip": 119092,
    "Kotak_Bluechip": 120841,
}

def fetch_scheme_nav(amfi_code, timeout=10):
    url = f"https://api.mfapi.in/mf/{amfi_code}"
    resp = requests.get(url, timeout=timeout)
    resp.raise_for_status()
    return resp.json()

def json_to_dataframe(payload, amfi_code):
    meta = payload.get("meta", {})
    nav_data = payload.get("data", [])
    df = pd.DataFrame(nav_data)
    if not df.empty:
        df["date"] = pd.to_datetime(df["date"], format="%d-%m-%Y")
        df["nav"] = pd.to_numeric(df["nav"], errors="coerce")
        df = df.sort_values("date").reset_index(drop=True)
    df.insert(0, "amfi_code", amfi_code)
    df.insert(1, "scheme_name", meta.get("scheme_name", ""))
    df.insert(2, "fund_house", meta.get("fund_house", ""))
    return df

print("Functions defined for", len(SCHEMES), "schemes.")
```

```python
live_nav = {}
for label, code_ in SCHEMES.items():
    print(f"Fetching {label} ({code_}) ...")
    try:
        payload = fetch_scheme_nav(code_)
        df = json_to_dataframe(payload, code_)
        out_path = os.path.join(RAW_DIR, f"live_nav_{code_}_{label}.csv")
        df.to_csv(out_path, index=False)
        live_nav[label] = df
        print(f"  saved {len(df)} rows -> {out_path}")
    except requests.exceptions.RequestException as exc:
        print(f"  FAILED to fetch {label} ({code_}): {exc}")
    time.sleep(0.3)
```

```python
nav_history = dfs["nav_history"]

master_codes = set(fund_master['amfi_code'].unique())
nav_codes = set(nav_history['amfi_code'].unique())

missing_in_nav = sorted(master_codes - nav_codes)
extra_in_nav = sorted(nav_codes - master_codes)

summary = pd.DataFrame({
    "metric": [
        "schemes_in_fund_master",
        "schemes_in_nav_history",
        "schemes_missing_nav_data",
        "nav_codes_not_in_master",
        "nav_rows_total",
        "nav_date_min",
        "nav_date_max",
    ],
    "value": [
        len(master_codes),
        len(nav_codes),
        len(missing_in_nav),
        len(extra_in_nav),
        len(nav_history),
        nav_history['date'].min(),
        nav_history['date'].max(),
    ],
})

display(summary)

if missing_in_nav:
    print(f"WARNING: {len(missing_in_nav)} scheme(s) in fund_master have no nav_history rows:")
    print(missing_in_nav[:20])
else:
    print("OK: every amfi_code in fund_master has matching nav_history data.")

if extra_in_nav:
    print(f"NOTE: {len(extra_in_nav)} amfi_code(s) in nav_history are not in fund_master:")
    print(extra_in_nav[:20])
```
